```python
import torch
import os
import pandas as pd
import numpy as np
```

```python
tensor_dir = './BatchedTensors'
file_paths = os.listdir(tensor_dir)

loaded_tensors = []
# Load tensors from files
for path in file_paths:
    loaded_tensor = torch.load(f"{tensor_dir}/{path}")
    loaded_tensors.append(loaded_tensor)
```

```python
combined_tensor = torch.cat(loaded_tensors, dim = 0)
combined_tensor
```

```python
partner_headlines = pd.read_csv('./archive/raw_partner_headlines.csv')
benzinga_headlines = pd.read_csv('./archive/raw_analyst_ratings.csv')
headlines = pd.concat([partner_headlines, benzinga_headlines]).drop('Unnamed: 0', axis = 1)
headlines
```

```python
sp_100_data = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')
sp_100_cmpys = sp_100_data[2]
sp_100 = sp_100_cmpys['Symbol'].to_list()
sp_100
```

```python
sp_100 = headlines.loc[headlines['stock'].isin(sp_100)]
sp_100
```

```python
positive = combined_tensor[:, 0].tolist()
negative = combined_tensor[:, 1].tolist()
neutral = combined_tensor[:, 2].tolist()


table = {'Headline':sp_100['headline'].to_list(),
         'Ticker': sp_100['stock'].to_list(),
         'Date': sp_100['date'].to_list(),
         "Positive":positive,
         "Negative":negative, 
         "Neutral":neutral}
      
df = pd.DataFrame(table, columns = ["Headline", "Ticker", "Date", "Positive", "Negative", "Neutral"])
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
df
```

```python
df['Tone'] = df[['Positive', 'Negative', 'Neutral']].idxmax(axis = 1).replace({'Negative': -1, "Positive": 1, "Neutral": 0})
df
```

```python
df['sentiment_embedding'] = np.tanh((np.log(df['Positive'] / df['Negative'])/ df['Neutral']))
df
```

```python
means_df = df.groupby(['Ticker', 'Date']).agg({'Tone': 'mean', 'sentiment_embedding': 'mean'}).reset_index()
means_df
```

```python
means_df.to_csv('./news_sentiment_data.csv')
```
